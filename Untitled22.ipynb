```python
# We now prepare the CIFAKE images for the baseline CNN model.
# CIFAKE images are originally 32 by 32 pixels, so we keep this size for the first baseline experiment.
# The training set is split into training and validation subsets so we can monitor overfitting.

baseline_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.5, 0.5, 0.5],
        std=[0.5, 0.5, 0.5]
    )
])

full_train_dataset = datasets.ImageFolder(
    root=train_dir,
    transform=baseline_transform
)

test_dataset = datasets.ImageFolder(
    root=test_dir,
    transform=baseline_transform
)

class_names = full_train_dataset.classes
num_classes = len(class_names)

train_size = int(0.8 * len(full_train_dataset))
val_size = len(full_train_dataset) - train_size

train_dataset, val_dataset = random_split(
    full_train_dataset,
    [train_size, val_size],
    generator=torch.Generator().manual_seed(SEED)
)

batch_size = 256

train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

val_loader = DataLoader(
    val_dataset,
    batch_size=batch_size,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=batch_size,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

print("Classes:", class_names)
print("Number of classes:", num_classes)
print("Training samples:", len(train_dataset))
print("Validation samples:", len(val_dataset))
print("Testing samples:", len(test_dataset))
print("Batch size:", batch_size)
```

```python
# We check one batch from the training DataLoader.
# This confirms that images and labels are being loaded in the correct tensor format.

images, labels = next(iter(train_loader))

print("Image batch shape:", images.shape)
print("Label batch shape:", labels.shape)
print("Minimum pixel value after normalization:", images.min().item())
print("Maximum pixel value after normalization:", images.max().item())
print("First 10 labels:", labels[:10].tolist())
```

```python
# We now define the baseline CNN model.
# This model is intentionally simple because it acts as the control group for the project.
# Later, we will compare it with stronger pretrained models such as MobileNetV2, ResNet50, and EfficientNetB0.

class BaselineCNN(nn.Module):
    def __init__(self, num_classes=2):
        super(BaselineCNN, self).__init__()

        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),

            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 4 * 4, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x


baseline_model = BaselineCNN(num_classes=num_classes).to(device)

total_params = sum(p.numel() for p in baseline_model.parameters())
trainable_params = sum(p.numel() for p in baseline_model.parameters() if p.requires_grad)

print(baseline_model)
print("\nTotal parameters:", total_params)
print("Trainable parameters:", trainable_params)
```

```python
# We now create reusable training and evaluation functions.
# These functions will help us train each model consistently and compare results fairly.

def train_one_epoch(model, data_loader, criterion, optimizer, device):
    model.train()

    running_loss = 0.0
    all_predictions = []
    all_labels = []

    for images, labels in data_loader:
        images = images.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)

        optimizer.zero_grad()

        outputs = model(images)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item() * images.size(0)

        predictions = torch.argmax(outputs, dim=1)

        all_predictions.extend(predictions.detach().cpu().numpy())
        all_labels.extend(labels.detach().cpu().numpy())

    epoch_loss = running_loss / len(data_loader.dataset)
    epoch_accuracy = accuracy_score(all_labels, all_predictions)
    epoch_f1 = f1_score(all_labels, all_predictions, average="binary", pos_label=1)

    return epoch_loss, epoch_accuracy, epoch_f1


def evaluate_model(model, data_loader, criterion, device):
    model.eval()

    running_loss = 0.0
    all_predictions = []
    all_labels = []
    all_probabilities = []

    with torch.no_grad():
        for images, labels in data_loader:
            images = images.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)

            outputs = model(images)
            loss = criterion(outputs, labels)

            running_loss += loss.item() * images.size(0)

            probabilities = torch.softmax(outputs, dim=1)[:, 1]
            predictions = torch.argmax(outputs, dim=1)

            all_predictions.extend(predictions.detach().cpu().numpy())
            all_labels.extend(labels.detach().cpu().numpy())
            all_probabilities.extend(probabilities.detach().cpu().numpy())

    epoch_loss = running_loss / len(data_loader.dataset)
    epoch_accuracy = accuracy_score(all_labels, all_predictions)
    epoch_precision = precision_score(all_labels, all_predictions, average="binary", pos_label=1)
    epoch_recall = recall_score(all_labels, all_predictions, average="binary", pos_label=1)
    epoch_f1 = f1_score(all_labels, all_predictions, average="binary", pos_label=1)

    try:
        epoch_auc = roc_auc_score(all_labels, all_probabilities)
    except ValueError:
        epoch_auc = None

    return {
        "loss": epoch_loss,
        "accuracy": epoch_accuracy,
        "precision": epoch_precision,
        "recall": epoch_recall,
        "f1": epoch_f1,
        "auc": epoch_auc,
        "labels": all_labels,
        "predictions": all_predictions,
        "probabilities": all_probabilities
    }


def train_model(model, train_loader, val_loader, criterion, optimizer, device, epochs, model_name):
    history = {
        "epoch": [],
        "train_loss": [],
        "train_accuracy": [],
        "train_f1": [],
        "val_loss": [],
        "val_accuracy": [],
        "val_precision": [],
        "val_recall": [],
        "val_f1": [],
        "val_auc": []
    }

    best_val_f1 = 0.0
    best_model_path = MODELS_DIR / f"{model_name}_best.pth"

    for epoch in range(1, epochs + 1):
        start_time = time.time()

        train_loss, train_accuracy, train_f1 = train_one_epoch(
            model,
            train_loader,
            criterion,
            optimizer,
            device
        )

        val_results = evaluate_model(
            model,
            val_loader,
            criterion,
            device
        )

        epoch_time = time.time() - start_time

        history["epoch"].append(epoch)
        history["train_loss"].append(train_loss)
        history["train_accuracy"].append(train_accuracy)
        history["train_f1"].append(train_f1)
        history["val_loss"].append(val_results["loss"])
        history["val_accuracy"].append(val_results["accuracy"])
        history["val_precision"].append(val_results["precision"])
        history["val_recall"].append(val_results["recall"])
        history["val_f1"].append(val_results["f1"])
        history["val_auc"].append(val_results["auc"])

        if val_results["f1"] > best_val_f1:
            best_val_f1 = val_results["f1"]
            torch.save(model.state_dict(), best_model_path)

        print(
            f"Epoch {epoch}/{epochs} | "
            f"Train Loss: {train_loss:.4f} | "
            f"Train Acc: {train_accuracy:.4f} | "
            f"Train F1: {train_f1:.4f} | "
            f"Val Loss: {val_results['loss']:.4f} | "
            f"Val Acc: {val_results['accuracy']:.4f} | "
            f"Val F1: {val_results['f1']:.4f} | "
            f"Val AUC: {val_results['auc']:.4f} | "
            f"Time: {epoch_time:.2f}s"
        )

    history_df = pd.DataFrame(history)
    history_path = RESULTS_DIR / f"{model_name}_training_history.csv"
    history_df.to_csv(history_path, index=False)

    print("\nBest validation F1:", best_val_f1)
    print("Best model saved to:", best_model_path)
    print("Training history saved to:", history_path)

    return history_df, best_model_path
```

```python
# We now train the Vanilla CNN baseline.
# This gives us the first performance result that later models must improve upon.

criterion = nn.CrossEntropyLoss()

baseline_optimizer = torch.optim.Adam(
    baseline_model.parameters(),
    lr=0.001,
    weight_decay=1e-4
)

baseline_epochs = 10

baseline_history, baseline_best_path = train_model(
    model=baseline_model,
    train_loader=train_loader,
    val_loader=val_loader,
    criterion=criterion,
    optimizer=baseline_optimizer,
    device=device,
    epochs=baseline_epochs,
    model_name="baseline_cnn"
)
```

```python
# We now evaluate the best saved Vanilla CNN model on the separate test set.
# This test result becomes the official baseline result for the project.

best_baseline_model = BaselineCNN(num_classes=num_classes).to(device)
best_baseline_model.load_state_dict(torch.load(baseline_best_path, map_location=device))

baseline_test_results = evaluate_model(
    model=best_baseline_model,
    data_loader=test_loader,
    criterion=criterion,
    device=device
)

print("Baseline CNN Test Results")
print("Test Loss:", round(baseline_test_results["loss"], 4))
print("Test Accuracy:", round(baseline_test_results["accuracy"], 4))
print("Test Precision:", round(baseline_test_results["precision"], 4))
print("Test Recall:", round(baseline_test_results["recall"], 4))
print("Test F1-score:", round(baseline_test_results["f1"], 4))
print("Test ROC-AUC:", round(baseline_test_results["auc"], 4))
```

```python
# We now create the confusion matrix and classification report.
# This helps us understand whether the model is confusing REAL images as FAKE or FAKE images as REAL.

baseline_cm = confusion_matrix(
    baseline_test_results["labels"],
    baseline_test_results["predictions"]
)

print("Class order:", class_names)
print("\nConfusion Matrix:")
print(baseline_cm)

print("\nClassification Report:")
print(
    classification_report(
        baseline_test_results["labels"],
        baseline_test_results["predictions"],
        target_names=class_names
    )
)

plt.figure(figsize=(5, 4))
plt.imshow(baseline_cm)
plt.title("Baseline CNN Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.xticks(ticks=[0, 1], labels=class_names)
plt.yticks(ticks=[0, 1], labels=class_names)

for i in range(baseline_cm.shape[0]):
    for j in range(baseline_cm.shape[1]):
        plt.text(j, i, baseline_cm[i, j], ha="center", va="center")

plt.tight_layout()
plt.show()
```

```python
# We save the baseline CNN test results into a results table.
# Later, we will append results from MobileNetV2, ResNet50, and EfficientNetB0 for final comparison.

model_results = []

model_results.append({
    "model": "Vanilla CNN",
    "test_loss": baseline_test_results["loss"],
    "test_accuracy": baseline_test_results["accuracy"],
    "test_precision": baseline_test_results["precision"],
    "test_recall": baseline_test_results["recall"],
    "test_f1": baseline_test_results["f1"],
    "test_auc": baseline_test_results["auc"],
    "parameters": sum(p.numel() for p in best_baseline_model.parameters()),
    "trainable_parameters": sum(p.numel() for p in best_baseline_model.parameters() if p.requires_grad)
})

baseline_results_df = pd.DataFrame(model_results)
baseline_results_path = RESULTS_DIR / "clean_test_results.csv"
baseline_results_df.to_csv(baseline_results_path, index=False)

baseline_results_df
```

```python
# We now prepare a second set of DataLoaders for pretrained transfer-learning models.
# These models were originally trained on ImageNet, so the images are resized to 224 by 224
# and normalized using ImageNet statistics.

pretrained_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

full_train_dataset_224 = datasets.ImageFolder(
    root=train_dir,
    transform=pretrained_transform
)

test_dataset_224 = datasets.ImageFolder(
    root=test_dir,
    transform=pretrained_transform
)

train_size_224 = int(0.8 * len(full_train_dataset_224))
val_size_224 = len(full_train_dataset_224) - train_size_224

train_dataset_224, val_dataset_224 = random_split(
    full_train_dataset_224,
    [train_size_224, val_size_224],
    generator=torch.Generator().manual_seed(SEED)
)

pretrained_batch_size = 128

train_loader_224 = DataLoader(
    train_dataset_224,
    batch_size=pretrained_batch_size,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

val_loader_224 = DataLoader(
    val_dataset_224,
    batch_size=pretrained_batch_size,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

test_loader_224 = DataLoader(
    test_dataset_224,
    batch_size=pretrained_batch_size,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

images_224, labels_224 = next(iter(train_loader_224))

print("Classes:", full_train_dataset_224.classes)
print("Training samples:", len(train_dataset_224))
print("Validation samples:", len(val_dataset_224))
print("Testing samples:", len(test_dataset_224))
print("Batch size:", pretrained_batch_size)
print("Image batch shape:", images_224.shape)
print("Label batch shape:", labels_224.shape)
```

```python
# We now define the MobileNetV2 transfer-learning model.
# MobileNetV2 is lightweight, so it is useful for cybersecurity applications where fast image screening is important.

mobilenet_model = models.mobilenet_v2(
    weights=models.MobileNet_V2_Weights.DEFAULT
)

in_features = mobilenet_model.classifier[1].in_features

mobilenet_model.classifier[1] = nn.Linear(
    in_features,
    num_classes
)

mobilenet_model = mobilenet_model.to(device)

mobilenet_total_params = sum(p.numel() for p in mobilenet_model.parameters())
mobilenet_trainable_params = sum(p.numel() for p in mobilenet_model.parameters() if p.requires_grad)

print(mobilenet_model.classifier)
print("Total parameters:", mobilenet_total_params)
print("Trainable parameters:", mobilenet_trainable_params)
```

```python
# We now train MobileNetV2.
# This model represents the lightweight transfer-learning approach in our project.

mobilenet_optimizer = torch.optim.Adam(
    mobilenet_model.parameters(),
    lr=0.0001,
    weight_decay=1e-4
)

mobilenet_epochs = 5

mobilenet_history, mobilenet_best_path = train_model(
    model=mobilenet_model,
    train_loader=train_loader_224,
    val_loader=val_loader_224,
    criterion=criterion,
    optimizer=mobilenet_optimizer,
    device=device,
    epochs=mobilenet_epochs,
    model_name="mobilenetv2"
)
```

```python
# We now evaluate the best saved MobileNetV2 model on the separate test set.
# This result shows whether the lightweight transfer-learning model improves over the Vanilla CNN baseline.

best_mobilenet_model = models.mobilenet_v2(
    weights=None
)

in_features = best_mobilenet_model.classifier[1].in_features
best_mobilenet_model.classifier[1] = nn.Linear(
    in_features,
    num_classes
)

best_mobilenet_model = best_mobilenet_model.to(device)
best_mobilenet_model.load_state_dict(torch.load(mobilenet_best_path, map_location=device))

mobilenet_test_results = evaluate_model(
    model=best_mobilenet_model,
    data_loader=test_loader_224,
    criterion=criterion,
    device=device
)

print("MobileNetV2 Test Results")
print("Test Loss:", round(mobilenet_test_results["loss"], 4))
print("Test Accuracy:", round(mobilenet_test_results["accuracy"], 4))
print("Test Precision:", round(mobilenet_test_results["precision"], 4))
print("Test Recall:", round(mobilenet_test_results["recall"], 4))
print("Test F1-score:", round(mobilenet_test_results["f1"], 4))
print("Test ROC-AUC:", round(mobilenet_test_results["auc"], 4))
```

```python
# We now create the MobileNetV2 confusion matrix and classification report.
# This helps us compare the lightweight model with the Vanilla CNN baseline.

mobilenet_cm = confusion_matrix(
    mobilenet_test_results["labels"],
    mobilenet_test_results["predictions"]
)

print("Class order:", class_names)
print("\nConfusion Matrix:")
print(mobilenet_cm)

print("\nClassification Report:")
print(
    classification_report(
        mobilenet_test_results["labels"],
        mobilenet_test_results["predictions"],
        target_names=class_names
    )
)

plt.figure(figsize=(5, 4))
plt.imshow(mobilenet_cm)
plt.title("MobileNetV2 Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.xticks(ticks=[0, 1], labels=class_names)
plt.yticks(ticks=[0, 1], labels=class_names)

for i in range(mobilenet_cm.shape[0]):
    for j in range(mobilenet_cm.shape[1]):
        plt.text(j, i, mobilenet_cm[i, j], ha="center", va="center")

plt.tight_layout()
plt.show()
```

```python
# We now prepare transformed versions of the test dataset.
# These transformations simulate cybersecurity-relevant conditions such as image compression,
# blur, noise, resizing, and brightness changes.

from PIL import ImageFilter, ImageEnhance
import io

class JPEGCompressionTransform:
    def __init__(self, quality=35):
        self.quality = quality

    def __call__(self, image):
        buffer = io.BytesIO()
        image.save(buffer, format="JPEG", quality=self.quality)
        buffer.seek(0)
        return Image.open(buffer).convert("RGB")


class GaussianBlurTransform:
    def __init__(self, radius=1.5):
        self.radius = radius

    def __call__(self, image):
        return image.filter(ImageFilter.GaussianBlur(radius=self.radius))


class GaussianNoiseTransform:
    def __init__(self, mean=0.0, std=18.0):
        self.mean = mean
        self.std = std

    def __call__(self, image):
        image_array = np.array(image).astype(np.float32)
        noise = np.random.normal(self.mean, self.std, image_array.shape)
        noisy_image = image_array + noise
        noisy_image = np.clip(noisy_image, 0, 255).astype(np.uint8)
        return Image.fromarray(noisy_image)


class DownscaleUpscaleTransform:
    def __init__(self, small_size=64, final_size=224):
        self.small_size = small_size
        self.final_size = final_size

    def __call__(self, image):
        image = image.resize((self.small_size, self.small_size), Image.BILINEAR)
        image = image.resize((self.final_size, self.final_size), Image.BILINEAR)
        return image


class BrightnessTransform:
    def __init__(self, factor=0.55):
        self.factor = factor

    def __call__(self, image):
        enhancer = ImageEnhance.Brightness(image)
        return enhancer.enhance(self.factor)


def build_test_loader_with_transform(extra_transform, batch_size=128):
    transform_steps = []

    transform_steps.append(transforms.Resize((224, 224)))

    if extra_transform is not None:
        transform_steps.append(extra_transform)

    transform_steps.extend([
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225]
        )
    ])

    composed_transform = transforms.Compose(transform_steps)

    transformed_dataset = datasets.ImageFolder(
        root=test_dir,
        transform=composed_transform
    )

    transformed_loader = DataLoader(
        transformed_dataset,
        batch_size=batch_size,
        shuffle=False,
        num_workers=2,
        pin_memory=True
    )

    return transformed_loader


robustness_test_loaders = {
    "clean": test_loader_224,
    "jpeg_compression": build_test_loader_with_transform(JPEGCompressionTransform(quality=35)),
    "gaussian_blur": build_test_loader_with_transform(GaussianBlurTransform(radius=1.5)),
    "gaussian_noise": build_test_loader_with_transform(GaussianNoiseTransform(std=18.0)),
    "downscale_upscale": build_test_loader_with_transform(DownscaleUpscaleTransform(small_size=64, final_size=224)),
    "brightness_low": build_test_loader_with_transform(BrightnessTransform(factor=0.55))
}

print("Robustness test loaders created:")
for name, loader in robustness_test_loaders.items():
    images, labels = next(iter(loader))
    print(name, "->", images.shape, labels.shape)
```

```python
# We now evaluate MobileNetV2 and EfficientNetB0 under each robustness condition.
# This helps us understand which model is more reliable when images are compressed,
# blurred, resized, darkened, or affected by noise.

robustness_models = {
    "MobileNetV2": best_mobilenet_model,
    "EfficientNetB0": best_efficientnet_model
}

robustness_results = []

for model_name, model in robustness_models.items():
    print("\nEvaluating:", model_name)

    for condition_name, loader in robustness_test_loaders.items():
        results = evaluate_model(
            model=model,
            data_loader=loader,
            criterion=criterion,
            device=device
        )

        robustness_results.append({
            "model": model_name,
            "condition": condition_name,
            "loss": results["loss"],
            "accuracy": results["accuracy"],
            "precision": results["precision"],
            "recall": results["recall"],
            "f1": results["f1"],
            "auc": results["auc"]
        })

        print(
            f"{condition_name} | "
            f"Acc: {results['accuracy']:.4f} | "
            f"F1: {results['f1']:.4f} | "
            f"AUC: {results['auc']:.4f}"
        )

robustness_results_df = pd.DataFrame(robustness_results)
robustness_results_path = RESULTS_DIR / "robustness_results.csv"
robustness_results_df.to_csv(robustness_results_path, index=False)

robustness_results_df
```

```python
# We now define a training-time random Gaussian noise transform.
# This helps the improved model learn more stable features instead of relying only on clean-image artifacts.

class RandomGaussianNoise:
    def __init__(self, mean=0.0, std=0.03, probability=0.3):
        self.mean = mean
        self.std = std
        self.probability = probability

    def __call__(self, tensor):
        if random.random() < self.probability:
            noise = torch.randn_like(tensor) * self.std + self.mean
            tensor = tensor + noise
            tensor = torch.clamp(tensor, 0.0, 1.0)
        return tensor
```

```python
# We now create a robustness-aware training transform.
# The training images are randomly modified using crop, flip, blur, color jitter, and mild noise.
# This encourages the model to learn features that are more stable under real-world image transformations.

robust_train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomApply([
        transforms.GaussianBlur(kernel_size=3, sigma=(0.1, 1.0))
    ], p=0.25),
    transforms.ColorJitter(
        brightness=0.25,
        contrast=0.25,
        saturation=0.15,
        hue=0.03
    ),
    transforms.ToTensor(),
    RandomGaussianNoise(std=0.03, probability=0.3),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

robust_eval_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

full_train_dataset_robust = datasets.ImageFolder(
    root=train_dir,
    transform=robust_train_transform
)

full_train_dataset_robust_eval_split = datasets.ImageFolder(
    root=train_dir,
    transform=robust_eval_transform
)

test_dataset_robust_eval = datasets.ImageFolder(
    root=test_dir,
    transform=robust_eval_transform
)

train_indices = train_dataset_224.indices
val_indices = val_dataset_224.indices

from torch.utils.data import Subset

robust_train_dataset = Subset(full_train_dataset_robust, train_indices)
robust_val_dataset = Subset(full_train_dataset_robust_eval_split, val_indices)

robust_train_loader = DataLoader(
    robust_train_dataset,
    batch_size=128,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

robust_val_loader = DataLoader(
    robust_val_dataset,
    batch_size=128,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

robust_test_loader = DataLoader(
    test_dataset_robust_eval,
    batch_size=128,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

images_robust, labels_robust = next(iter(robust_train_loader))

print("Robust training samples:", len(robust_train_dataset))
print("Robust validation samples:", len(robust_val_dataset))
print("Robust testing samples:", len(test_dataset_robust_eval))
print("Robust training batch shape:", images_robust.shape)
print("Labels batch shape:", labels_robust.shape)
```

```python
# We now train a robustness-aware EfficientNetB0 model.
# This model is trained with random visual perturbations so it can become more reliable under real-world image changes.

robust_efficientnet_model = models.efficientnet_b0(
    weights=models.EfficientNet_B0_Weights.DEFAULT
)

in_features = robust_efficientnet_model.classifier[1].in_features
robust_efficientnet_model.classifier[1] = nn.Linear(
    in_features,
    num_classes
)

robust_efficientnet_model = robust_efficientnet_model.to(device)

robust_efficientnet_optimizer = torch.optim.Adam(
    robust_efficientnet_model.parameters(),
    lr=0.0001,
    weight_decay=1e-4
)

robust_efficientnet_epochs = 5

robust_efficientnet_history, robust_efficientnet_best_path = train_model(
    model=robust_efficientnet_model,
    train_loader=robust_train_loader,
    val_loader=robust_val_loader,
    criterion=criterion,
    optimizer=robust_efficientnet_optimizer,
    device=device,
    epochs=robust_efficientnet_epochs,
    model_name="robust_efficientnet_b0"
)
```

```python
# We now evaluate the best robustness-aware EfficientNetB0 model on the clean test set.
# This shows whether the augmented training preserved strong clean-image detection performance.

best_robust_efficientnet_model = models.efficientnet_b0(
    weights=None
)

in_features = best_robust_efficientnet_model.classifier[1].in_features
best_robust_efficientnet_model.classifier[1] = nn.Linear(
    in_features,
    num_classes
)

best_robust_efficientnet_model = best_robust_efficientnet_model.to(device)
best_robust_efficientnet_model.load_state_dict(
    torch.load(robust_efficientnet_best_path, map_location=device)
)

robust_efficientnet_test_results = evaluate_model(
    model=best_robust_efficientnet_model,
    data_loader=test_loader_224,
    criterion=criterion,
    device=device
)

print("Robust EfficientNetB0 Clean Test Results")
print("Test Loss:", round(robust_efficientnet_test_results["loss"], 4))
print("Test Accuracy:", round(robust_efficientnet_test_results["accuracy"], 4))
print("Test Precision:", round(robust_efficientnet_test_results["precision"], 4))
print("Test Recall:", round(robust_efficientnet_test_results["recall"], 4))
print("Test F1-score:", round(robust_efficientnet_test_results["f1"], 4))
print("Test ROC-AUC:", round(robust_efficientnet_test_results["auc"], 4))
```
